```python
def find_missing_columns(missing_df):
  missing_columns= []
  for column in missing_df.columns:
    if True in missing_df[column].unique():
      missing_columns.append(column)
  return missing_columns
```

```python
def show_missing_column(df, missing_df, missing_columns):
  for column in missing_columns:
      print(f"""
            {column}
            {missing_df[column].value_counts().to_dict()}
            {df[column].unique()}
            {df[column].unique().dtype}
            {'-' * 70}
          """)
```

## Data Normalization

```python
def normalization_simple_feature_scaling(feature):
  print(f'before normalization: {df[feature].head()}')
  df[feature] = df[feature] / df[feature].max()
  print(f'after normalization: {df[feature].head()}')
```

## Binning

```python
def bin_feature(feature, num_bins, labels):
  min = df[feature].min()
  max = df[feature].max()
  bins = np.linspace(min, max, num_bins)

  feature_binned = f'{feature}-binned'
  df[feature_binned] = pd.cut(df[feature], bins=bins, labels=labels, include_lowest=True)
  return df[[feature, feature_binned]]
```

```python
bin_feature('horsepower', 4, ['Low', 'Medium', 'High'])
```

## Convert categorical var to dummy var

```python
def convert_cat_num(feature, df):
  dummies_df = pd.get_dummies(df[feature])
  dummies_var = dummies_df.columns

  mapping_dummies_var = {}
  for dummy_var in dummies_var:
    key = dummy_var
    value = f'{feature}-{dummy_var}'
    mapping_dummies_var[key] = value

  dummies_df.rename(columns=mapping_dummies_var, inplace=True)

  df = pd.concat([df, dummies_df], axis=1)
  df.drop(feature, axis=1, inplace=True)
  return df
```

```python
df = convert_cat_num('aspiration', df)
```
