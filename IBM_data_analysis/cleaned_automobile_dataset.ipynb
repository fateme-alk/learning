## Missing values

```python
df.replace("?", np.nan, inplace=True)
```

```python
df
```

```python
missing_df = df.isnull()
```

```python
missing_df
```

```python
def find_missing_columns(missing_df):
  missing_columns= []
  for column in missing_df.columns:
    if True in missing_df[column].unique():
      missing_columns.append(column)
  return missing_columns
```

```python
def show_missing_column(df, missing_df, missing_columns):
  for column in missing_columns:
      print(f"""
            {column}
            {missing_df[column].value_counts().to_dict()}
            {df[column].unique()}
            {df[column].unique().dtype}
            {'-' * 70}
          """)
```

```python
missing_columns = find_missing_columns(missing_df)
```

```python
missing_columns
```

```python
show_missing_column(df, missing_df, missing_columns)
```

```python
def fillna_with_mean(column):
  avg = df[column].astype("float").mean(skipna=True)
  df.fillna({column: avg}, inplace=True)
  print(f'mean of {column} column is: {avg} \nThe dtype is: {df[column].unique().dtype}')
```

```python
fillna_with_mean("normalized-losses")
```

```python
fillna_with_mean("bore")
```

```python
fillna_with_mean("stroke")
```

```python
fillna_with_mean("horsepower")
```

```python
fillna_with_mean("peak-rpm")
```

```python
def fillna_with_mode(column):
  mode = df[column].mode(dropna=True).to_dict()[0]
  # we can use below line of code instead of above line too
  # df['num-of-doors'].value_counts().idxmax()
  df.fillna({column: mode}, inplace=True)
  print(f'mode of {column} column is: {mode} \nThe dtype is: {df[column].unique().dtype}')
```

```python
fillna_with_mode("num-of-doors")
```

## Data Normalization

```python
def normalization_simple_feature_scaling(feature):
  print(f'before normalization: {df[feature].head()}')
  df[feature] = df[feature] / df[feature].max()
  print(f'after normalization: {df[feature].head()}')
```

```python
normalization_simple_feature_scaling('width')
```

```python
normalization_simple_feature_scaling('height')
```
