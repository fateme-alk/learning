## statistics for a numerical feature




```python
print(df['symboling'].count())
# .mean()excludes np.NaN values when calculating the average. Setting skipna=True (which is the default) will exclude np.NaN values, while setting skipna=False will raise an error if any np.NaN values are present.
print(df['symboling'].mean())
print(df['symboling'].std())
print(df['symboling'].min())
print(df['symboling'].max())
print(df['symboling'].quantile(0.25))
print(df['symboling'].quantile(0.5))
print(df['symboling'].quantile(0.75))

# The .unique() method is applicable to all columns, regardless of their data type. It returns an array of the unique values in the column.
print(df['symboling'].unique())

# Below line will raise error regardless of data type, we have to use mode() method instead like below line
# print(df['symboling'].top())
print(df['symboling'].mode())

# The freq() method does not exist in pandas. If you try to call it, you will get an error
# print(df['symboling'].freq())
```
